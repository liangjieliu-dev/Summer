```python
import pandas as pd
from collections import Counter

file_path =  '一列分N列不重复.xlsm'
df = pd.read_excel(file_path,header=None)
df.columns =['staffName','None','None']
# 清理数据中的换行符
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
names = df.iloc[:,0]
```

```python
staff_dict = {}
for name in names:
    if name in staff_dict:
        staff_dict[name].append(name)
    else:
        staff_dict[name] = [name]
```

```python
# 函数：从数量最多的前n个堆中取出一个人
def take_staffs(staff_dict, number):
    staff_counts = {k: len(v) for k, v in staff_dict.items()}
    valuesum = 0
    for value in staff_counts.values():
        valuesum = valuesum + value
    print(valuesum)
    # 找出数量最多的前n个堆
    top_counts = Counter(staff_counts).most_common(number)
    print(top_counts)
    taken_staffs = []
    
    # 从每个堆中取出一个人
    for name, _ in top_counts:
        if staff_dict[name]:
            taken_staffs.append(staff_dict[name].pop(0))
    print('taken_staffs:', len(taken_staffs))
    return taken_staffs
```

```python
classes=43
iteration=65

columns = [f"取出第{i+1}次" for i in range(iteration)]  
matrix = pd.DataFrame(columns=columns)
for i in range(iteration):
    taken_staffs = take_staffs(staff_dict,classes)
    matrix[columns[i]] = pd.Series(taken_staffs)
output_file = 'out.xlsx'
matrix.to_excel(output_file, index=False)
```
